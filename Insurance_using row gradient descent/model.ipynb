```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('TravelInsurancePrediction.csv')
df.head()
```

```python
data = df[["Age","TravelInsurance","ChronicDiseases"]]
data.head()
```

```python
data.isnull().sum()
```

```python
x=data[["Age","ChronicDiseases"]]
y=data["TravelInsurance"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train_scaled = x_train.copy()
x_train_scaled["Age"]=x_train["Age"]/100

x_test_scaled = x_test.copy()
x_test_scaled["Age"]=x_test["Age"]/100
```

```python
x_test_scaled
```

```python
model = keras.models.Sequential([
    keras.layers.Dense(1,activation="sigmoid",kernel_initializer='ones',bias_initializer='zeros')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_scaled,y_train,epochs=100)
```

```python
model.evaluate(x_test_scaled,y_test)
```

```python
def log_loss(y_true,y_pred):
    epsilon=1e-8
    y_predication = [max(i,epsilon) for i in y_pred ]
    y_predication_new = [min(i,1-epsilon) for i in y_predication ]
    return -np.mean(y_true*np.log(y_predication_new) + (1-y_true)*np.log(y_predication_new))
```

```python
import math
def sigmoid_array(arg):
    return 1/(1+np.exp(-arg))
```

```python
def gradient_descent(age,disease,y_true,iterations):
    w1=w2=1
    bias=0
    rate=0.5
    n=len(age)
    for i in range(iterations):
        weighted_sum = w1*age + w2*disease + bias
        y_pred = sigmoid_array(weighted_sum)
        loss = log_loss(y_true, y_pred)
        
        w1d = (1/n)*np.dot(np.transpose(age),(y_pred - y_true))
        w2d = (1/n)*np.dot(np.transpose(disease),(y_pred - y_true))
        bias_d = np.mean(y_pred - y_true)
        
        w1 = w1 - rate*w1d
        w2 = w2 - rate*w2d
        bias = bias - rate*bias_d
        
        # print(f"w1: {w1}, w2d: {w2}, bias: {bias} loss: {loss}")
        if loss <= 0.4:
            break
        
    return w1,w2,bias
        
    
```

```python
w1,w2,bias=gradient_descent(x_train_scaled["Age"],x_train_scaled["ChronicDiseases"],y_train,50)

def prediction_function_manual(ages,disease):
    weighted_sum = w1*ages + w2*disease + bias
    return sigmoid(weighted_sum)

prediction_function_manual(.23,1)
```
